What update anomalies occur in the EMP_PROJ and EMP_DEPT relations?

**EMP_DEPT**:

| Ename | Ssn | Bdate | Address | Dnumber | Dname | Dmgr_ssn |
| --- | --- | --- | --- | --- | --- | --- |
| - | PK | - | - | - | - | - |

The transitive dependency Ssn -> Dnumber ; Dnumber -> (Dname, Dmgr_ssn) can cause many issues:
* If a Department has no employees working for it (ie if we remove all employees working in that Department), it will lose it's representation within the database, since Ssn -> (Dname, Dmgr_ssn) due to the transitive property.
* A new Department may not be added unless an Employee works for it.
* Insert anomalies may occur if we try to insert a tuple relating a new Employee to an existing Department
    * All transitive dependencies must be checked before the database will accept the insert
        * For example, if we try to insert a tuple with a Dmgr_ssn value that is different than all existing Dmgr_ssn values for a given Dnumber, we will get an insert error
        * It is expensive to check all transitive dependencies each time we go to do an insert/update

**EMP_PROJ**

| Ssn | Pnumber | Hours | Ename | Pname | Plocation |
| --- | --- | --- | --- | --- | --- |
| PK | PK | - | - | - | - |

Ssn -> Ename<br>
Pnumber -> Pname, Plocation<br>

Partial dependencies cause many anomalies with this relation:
* Ssn -> Ename ; Pnumber -> (Pname, Plocation)
    * Let's say we have a Project with no Employees working on it
        * When the last Employee working on a project is removed, the Project will also be removed
    * A new Project cannot be added unless at least one Employee is assigned to work on it
    * Inserting a new tuple into this relation requires that we check both partial dependencies
    
We will also encounter update anomalies if we try to insert a tuple that has a different value for Plocation than the existing tuples with the same Pnumber value.

##### What is the key for R?

1) First we can apply the Decomposition or Projective property:<br>
$\left\{X \rightarrow YZ\right\} |= X \rightarrow Y$<br>
* Applying this rule repeatedly can decompose the FD $X \rightarrow \left\{A_{1}, A_{2}, \ldots , A_{n}\right\}$ into the set of dependencies $\left\{X \rightarrow A_{1}, X \rightarrow A_{2}, \ldots , X \rightarrow A_{n}\right\}$


$\left\{A\right\}\rightarrow\left\{D\right\}$<br>
$\left\{A\right\}\rightarrow\left\{E\right\}$<br>
$\left\{B\right\}\rightarrow\left\{F\right\}$<br>
$\left\{D\right\}\rightarrow\left\{I\right\}$<br>
$\left\{D\right\}\rightarrow\left\{J\right\}$<br>
$\left\{F\right\}\rightarrow\left\{G\right\}$<br>
$\left\{F\right\}\rightarrow\left\{H\right\}$<br>

2) Now we can apply the Transitive property:<br>
If $X \rightarrow Y$ and $Y \rightarrow Z$, then $X \rightarrow Z$<br>

$\left\{A\right\}\rightarrow\left\{D\right\}$<br>
And<br>
$\left\{D\right\}\rightarrow\left\{I\right\}$<br>

So...

$\left\{A\right\}\rightarrow\left\{I\right\}$<br>

and 

$\left\{A\right\}\rightarrow\left\{D\right\}$<br>
And<br>
$\left\{D\right\}\rightarrow\left\{J\right\}$<br>

So...

$\left\{A\right\}\rightarrow\left\{J\right\}$<br>

...

$\left\{B\right\}\rightarrow\left\{F\right\}$<br>
And<br>
$\left\{F\right\}\rightarrow\left\{G\right\}$<br>

So...

$\left\{B\right\}\rightarrow\left\{G\right\}$<br>

and

$\left\{B\right\}\rightarrow\left\{F\right\}$<br>
And<br>
$\left\{F\right\}\rightarrow\left\{H\right\}$<br>

So...

$\left\{B\right\}\rightarrow\left\{H\right\}$<br>

Collecting all of our results...

$\left\{A,B\right\}\rightarrow\left\{C\right\}$<br>
$\left\{A\right\}\rightarrow\left\{D\right\}$<br>
$\left\{A\right\}\rightarrow\left\{E\right\}$<br>
$\left\{B\right\}\rightarrow\left\{F\right\}$<br>
$\left\{B\right\}\rightarrow\left\{G\right\}$<br>
$\left\{B\right\}\rightarrow\left\{H\right\}$<br>
$\left\{A\right\}\rightarrow\left\{I\right\}$<br>
$\left\{A\right\}\rightarrow\left\{J\right\}$<br>

We see that the key $\left\{A,B\right\}$ satisfies all non-key attributes functional dependencies $\left\{C,D,E,F,G,H,I,J\right\}$<br>

##### Decompose into 2NF

To decompose into 2NF: $R$ is in 2NF iff $R$ is in 1NF and every nonkey attribute is fully dependent on the key.<br>
To acheive this, we just need to seperate out the relations based on FDs:<br>

**R1**:

R1(<u>A</u>,D,E,I,J)

R1 satisfies all FDs where $A$ is on the LHS of the arrow.

**R2**:

R2(<u>B</u>,F,G,H)

R2 satisfies all FDs where $B$ is on the LHS of the arrow.

**R3**:

R3(<u>A</u>,<u>B</u>,C)

R3 satisfies the case where $A$,$B$ is on the LHS of the arrow

$\left\{\text{R1, R2, R3}\right\}$ are in 2NF.

##### Decompose into 3NF

To decompose into 3NF: $R$ is in 3NF iff $R$ is in 2NF and every nonkey attribute is non-transitively dependent on the key.<br>
We must deal with the transitivity between

$\left\{A\right\}\rightarrow\left\{D,E\right\}$<br>
$\left\{D\right\}\rightarrow\left\{I,J\right\}$<br>

and 

$\left\{B\right\}\rightarrow\left\{F\right\}$<br>
$\left\{F\right\}\rightarrow\left\{G,H\right\}$<br>

Let's deal with

$\left\{A\right\}\rightarrow\left\{D,E\right\}$<br>
$\left\{D\right\}\rightarrow\left\{I,J\right\}$<br>

first by breaking relation R1 into two based on these FDs:

**R1a**:

R1a(<u>A</u>,D,E)

**R1b**:

R1b(<u>D</u>,I,J)
* R1b.D is a FK to R1a.D

We've removed the transitive dependencies from FDs where $A$ is on the LHS. Onward to the ones where $B$ is on the LHS!

$\left\{B\right\}\rightarrow\left\{F\right\}$<br>
$\left\{F\right\}\rightarrow\left\{G,H\right\}$<br>

We can break 

**R2**:

R2(<u>B</u>,F,G,H)

into the following relations based on the FDs above:

**R2a**:

R2a(<u>B</u>,F)

**R2b**:

R1b(<u>F</u>,G,H)
* R2b.F is a FK to R2a.F

Collectively, $\left\{\text{R1a, R1b, R2a, R2b, R3}\right\}$ are in 3NF.
